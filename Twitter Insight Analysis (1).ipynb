# Local and global thought patterns

```python
import json
import matplotlib.pyplot as plt
import pandas as pd
from textblob import TextBlob
from collections import Counter
```

```python
# Load WW_trends and US_trends data into the given variables respectively
with open(r"C:\Users\uttka\Desktop\Data\WWTrends.json") as f:
    W_Trends = json.load(f)
with open(r"C:\Users\uttka\Desktop\Data\USTrends.json") as f:
    US_Trends = json.load(f)

print(W_Trends)
print('='*140)
print(US_Trends)
```

# Prettifying the output

```python
print("W Trends:")
print(json.dumps(W_Trends,indent=1))

print("\nUS Trends")
print(json.dumps(US_Trends,indent=1))
```

# Finding common trends

```python
# Extracting all the W trend names from W_trends
world_trends = set([trend['name'] for trend in W_Trends[0]['trends']])
# Extracting all the US trend names from US_trends
us_trends = set([trend['name'] for trend in US_Trends[0]['trends']])
# Getting the intersection of the two sets of trends
common_trends = world_trends & us_trends

# Inspecting the data
print("WORLD TRENDS:",world_trends,"\n")
print("US TRENDS:",us_trends,"\n")
print(f"{len(common_trends)}common trends:", common_trends)
```

# Exploring the hot trend

```python
with open(r"C:\Users\uttka\Desktop\Data\WeLoveTheEarth.json") as f:
    tweets = json.load(f)
```

```python
# Inspecting some tweets
tweets[0:2]
```

# Performing Sentiment Analysis

```python
texts = [tweet['text'] for tweet in tweets]sentiments = [TextBlob(text).sentiment.polarity for text in texts]
# Inspecting the first 10 sentiment scores
print(json.dumps(sentiments[0:10], indent=1), "\n")
```

```python
# Plotting the distribution of sentiment scores
plt.figure(figsize=(10, 5))
plt.hist(sentiments, bins=20, color='blue', edgecolor='black')
plt.title('Sentiment Polarity Distribution')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```

# Digging deeper

```python
texts = [tweet['text'] for tweet in tweets]
# Extracting screen names of users tweeting about #WeLoveTheEarth
names = [user_mention['screen_name'] for tweet in tweets for user_mention in tweet['entities']['user_mentions']]

# Extracting all the hashtags being used when talking about this topic
hashtags = [hashtag['text'] for tweet in tweets for hashtag in tweet['entities']['hashtags']]


# Inspecting the first 10 results
print (json.dumps(texts[0:10], indent=1),"\n")
print (json.dumps(names[0:10], indent=1),"\n")
print (json.dumps(hashtags[0:10], indent=1),"\n")
```

# Frequency analysis

```python
# Counting occurrences/ getting frequency dist of all names and hashtags
for item in [names, hashtags]:
    c = Counter(item)
# Inspecting the 10 most common items in c
    print (c.most_common(10), "\n")
```

# Activity around the trend

```python
retweets = [(tweet['retweet_count'], tweet['retweeted_status']['favorite_count'], tweet['retweeted_status']['user']['followers_count'], tweet['retweeted_status']['user']['screen_name'], tweet['text']) for tweet in tweets if 'retweeted_status' in tweet]
```

# A table that speaks a 1000 words

```python
# Create a DataFrame and visualize the data in a pretty and insightful format
df = pd.DataFrame(retweets, columns=['Retweets', 'Favorites', 'Followers', 'ScreenName', 'Text']).groupby(['ScreenName','Text', 'Followers']).sum().sort_values(by = ['Followers'], ascending = False)
df.style.background_gradient()
```

# Analyzing used languages

```python
tweets_languages = [tweet['lang'] for tweet in tweets]
plt.figure(figsize=(10, 5))
plt.hist(tweets_languages, bins=len(set(tweets_languages)), color='blue', edgecolor='black')
plt.title('Language Distribution of Tweets')
plt.xlabel('Language')
plt.ylabel('Frequency')
plt.show()
    
```
